## PARTE I

En este ejercicio debes simular un chatbot al que le haremos pedidos de pizzas.

1. El chatbot tiene que saludar con un: "Buenas tardes, bienvenido al servicio de pedido online, el precio de nuestra pizzas es de 8,95 para la familiar o de 9,90 para la mediana ¿Cuántas pizzas  familiares desea?"
2. El ususario tiene que introducir un número de pizzas en una variable llamada 'pizz_familiar'
3. El chatbot debe responder: "Estupendo, ¿cuántas pizzas medianas desea?" y guardar el número de pizzas medianas en otra variable 'pizz_mediana'
4. Además debe responder del chatbot: "Estupendo, se están preparando 'pizz' pizzas familiares y 'pizz_mediana' pizzas medianas. Digame su dirección"
5. El ususario tiene que introducir una direccion en formato String en otra variable llamada 'direcc'
6. El programa debe calcular el monto total
7. Respuesta final del chatbot: "Le mandaremos las '(número total de pizzas)' pizzas a la dirección 'direcc'. Serán <precio total a pagar en euros>. Muchas gracias por su pedido."
         


# <font color="red" >Respuesta: </font>


````python
def exec_chatbot():
    pizz_familiar=input("Buenas tardes!!!,\n Bienvenido al servicio de pedido online, el precio de nuestra pizzas es de 8,95 para la familiar o de 9,90 para la mediana\n¿Cuántas pizzas  familiares desea?")
    pizz_familiar=int(pizz_familiar)
    pizz_mediana=input("Estupendo, ¿cuántas pizzas medianas desea?" )
    pizz_mediana=int(pizz_mediana)
    precio=round(8.95*pizz_mediana+9.90*pizz_familiar,3)
    total_pizzas=pizz_familiar+pizz_mediana
    direcc=input("Estupendo, se están preparando "+str(pizz_familiar)+" pizzas familiares y "+str(pizz_mediana)+" pizzas medianas. Digame su dirección")
    print("Le mandaremos las ",total_pizzas," pizzas a la dirección ",direcc,". Serán ",f"{precio:.2f}"," euros. Muchas gracias por su pedido!!!")
````

````python
exec_chatbot()
````

## Parte II

Crea un programa que realice las siguientes acciones:

1. Solicite al usuario que introduzca 3 números enteros separados por comas (por ejemplo: 1,2,3).
2. Convierta esa entrada en una lista de números enteros.
3. Utilizando funciones built-in, determine el número máximo, el número mínimo y la suma total de los números introducidos.
4. Determine la verdad o falsedad de la siguiente afirmación: "El número máximo menos el número mínimo es igual a la suma total dividida por 5".
5. Muestre los resultados utilizando print.

````python
def exec_programa():
    enteros=input("Introduzca 3 números enteros separados por comas (por ejemplo: 1,2,3)\n")
    lista=[int(i) for i in enteros.split(",")]
    assert len(lista)==3, "Error! El número de elementos en la lista hay que ser exactamente 3. Compuba si has introducido los numeros enteros correctamente."
    num_max=max(lista)
    num_min=min(lista)
    diferencia=num_max-num_min
    suma_dividida_por_5=sum(lista)/5
    result=diferencia==suma_dividida_por_5
    print("Lista:",lista,"\n")
    print("Num. max.:",num_max,"\n")
    print("Num. min.:",num_min,"\n")
    print("Diferencia entre max. y min.:",diferencia,"\n")
    print("Suma de elts en la lista dividida por 5:",suma_dividida_por_5,"\n")

    if result:
        print("El número máximo menos el número mínimo es igual a la suma total dividida por 5")
    else:
        print("El número máximo menos el número mínimo NO es igual a la suma total dividida por 5")
````

````python
exec_programa()
````
