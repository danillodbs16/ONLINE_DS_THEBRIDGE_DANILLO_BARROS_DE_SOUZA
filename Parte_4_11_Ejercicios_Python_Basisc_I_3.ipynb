## Ejercicio 3
Mismo programa que en 2, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

## <font color="red" >Respuesta para el ejercicio 3: </font>

````python
def son_iguales(A,B,C):
    output1=A==B==C
    output2=A==B or A==C or B==C
    
    if output1:
        print("Todos son iguales")
    else:
        if output2:
            print("Ni todos son iguales")
        else:
            print("Todos son distintos")
        
    return output1,output2
````

````python
son_iguales(4,4,4)
````

````python
son_iguales(1,0,1)
````

````python
son_iguales(12.4,0,1)
````

## Ejercicio 4
Razona sin ejecutar código el output que obtendremos de las siguientes sentencias
1. True and True and False
2. not ((True or False) and (True or False))
3. (False or False or False or True) and False
4. not False and True and not True

## <font color="red" >Respuesta para el ejercicio 4: </font>

1. Seguiendo la lógica matemática, las operaciones binarias (0's y 1's) son multiplicativas, así que True and True and False es equivalente a
$$  1 \times 1 \times 0 = 0$$.

O sea, que el output será False.

2. Por las propiedades de la lógica matemática, $\lnot (True \lor False) \land (True \lor False) =\lnot (True)  \land (True) = False \land True = False $.

3. Por las propiedades lógicas, $( False \lor False \lor False \lor True) \land False = (True) \land True = True$.

4. $ \lnot False \land True \land \lnot True = True \land True \land False = False$.
