## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

## <font color="red" >Respuesta para el ejercicio 2: </font>

```python
from math import pi
grados=input("Inserta aquí el grado que quieres convertir a radiales")
grados=float(grados)
radiales=grados*pi/180
print(grados," grados equivalen a ",radiales," radiales")
```

## Ejercicio 4
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

## <font color="red" >Respuesta para el ejercicio 4: </font>

```python
base=float(input("Introduzca la base: ").replace(",","."))
altura=float(input("Introduzca la altura: ").replace(",","."))
print("Área del paralelogramo: ",base*altura)
```
