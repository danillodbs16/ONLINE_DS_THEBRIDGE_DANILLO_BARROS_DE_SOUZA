### Definiendo la clase Tienda y sus atributos:

```python

class Tienda:
    """
    Clase que representa una tienda de electrodomésticos.
    
    Atributos comunes:
        tipo (str): Tipo de tienda (Electrodomésticos).
        abierta (bool): Indica si la tienda está abierta.
    
    Atributos propios:
        nombre (str): Nombre de la tienda.
        direcc (str): Dirección de la tienda.
        num_empleados (int): Número de empleados.
        ventas_ultimos_3_meses (list): Lista con las ventas de los últimos 3 meses.
    """

    def __init__(self,Nombre:str,Direccion:str,Num_empleados:int,Ventas_ultimos_meses:list):
        self.nombre=Nombre
        self.tipo="Eletrodomésticos"
        self.abierta=True
        self.direcc=Direccion
        self.num_empleados=Num_empleados
        self.ventas_ultimos_meses=Ventas_ultimos_meses
        
    def total_ventas(self):
        if self.num_empleados!=0:
            media=sum(self.ventas_ultimos_meses)/self.num_empleados
        else:
            media=0.
        return media
    def media_ventas(self):
        return sum(self.ventas_ultimos_meses)
        
    #def ventas_ultimo_mes(self):
     #   return self.ventas_ultimos_meses[-1]
        
    def proyeccion_ventas(self,X):
        ventas=total_ventas()
        if X<100:
            val=1.2
        else:
            val=1.5
        self.ventas_ultimos_meses=[val*i for i in self.ventas_ultimos_meses]
        return self.ventas_ultimos_meses
        
        
```

```python
Tienda1=Tienda("Tienda 1","Calle señor Fulano 0",9,[1000,1500,2000])
Tienda2=Tienda("Tienda 2","Avenida del diablo 666",666,[-666,-6666,-666666])
Tienda3=Tienda("Tienda 3","Calle de los Fallecidos 33",0,[0,0,0])
Tiendas=[Tienda1,Tienda2,Tienda3]
for tienda in Tiendas:
    print("Nombre de la Tienda:",tienda.nombre)
    print("Tipo:",tienda.tipo)
    print("Abierta:",tienda.abierta)
    print("Dirección:",tienda.direcc)
    print("Número de empleados:",tienda.num_empleados)
    print("Ventas últimos 3 meses:",tienda.ventas_ultimos_meses,"\n")
```

### Definiendo la clase Perro y sus atributos:

```python
class Perro:
    """
    Clase que representa un perro.

    Atributos:
        patas (int): Número de patas (4).
        orejas (int): Número de orejas (2).
        ojos (int): Número de ojos (2).
        raza (str): Raza del perro.
        pelo (str): Color del pelo (por defecto "Marrón").
        dueno (str or None): Nombre del dueño (por defecto None).
        velocidad (float): Velocidad actual del perro (por defecto 0).

    Métodos:
        andar(aumento_velocidad): Aumenta la velocidad del perro.
        parar(): Pone la velocidad a 0.
        ladrar(mensaje): Devuelve un string con "GUAU!" + mensaje.
    """
    def __init__(self,raza,velocidad=0):
        self.raza=raza
        self.velocidad=velocidad
        self.tiene_dueno=False
        self.tiene_4_patas=True
        self.tiene_2_orejas=True
        self.tiene_2_ojos=True
        self.color_pelo="marrón"
        
    def andar(self,aumento_velocidad):
        self.velocidad+=aumento_velocidad
    def parar(self):
        self.velocidad=0
    def ladrar(self,ladra:str=""):
        print("GUAU!",ladra)
        
        
```

```python
Perro1=Perro("Labrador")
print("Raza:",Perro1.raza)
print("velocidad:",Perro1.velocidad)
print("Tiene dueno:",Perro1.tiene_dueno)
print("Tiene 4 patas:",Perro1.tiene_4_patas)
print("Tiene 2 ojos:",Perro1.tiene_2_ojos)
print("Tiene 2 orejas:",Perro1.tiene_2_orejas)
print("Color de pelo:",Perro1.color_pelo)
```
