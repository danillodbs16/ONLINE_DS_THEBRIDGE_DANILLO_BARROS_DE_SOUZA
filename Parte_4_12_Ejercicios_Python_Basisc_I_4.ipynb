## Ejercicio 1

## <font color="red" >Respuesta para el ejercicio 1: </font>


```python
lista=[4,6,8,-1]
valor1=63.451256965
valor2=-74
print("Máximo de la lista: ",max(lista))
print("Suma de los elementos de la lista: ",sum(lista))
print("Redondeo del valor ",valor1," (3 dígitos): ",round(valor1,3))
print("Valor absoluto de ",valor2,": ",abs(valor2))
```

## Ejercicio 2
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."

## Ejercicio 3
1. Crea una lista con 3 elementos numéricos
2. Añade un cuarto elemento
3. Calcula la suma de todos
4. Elimina el segundo elemento de la lista
5. Añade otro elemento en la posicion 3 de la lista
6. Crea otra lista con 4 elementos y concaténala a la que ya tenías.
7. Ordena la lista de menor a mayor.
8. BONUS: Ordena la lista de mayor a menor.  
(*PISTA: Para hacer 8, tendrás que buscar en la información del método...*)  
  
  
NOTA: Pinta por pantalla el resultado de cada paso.

## <font color="red" >Respuesta para el ejercicio 2: </font>


```python
from math import pi,sqrt,log


#Creando una lista:
Lista=[10,-8,7.56]
print("1. Lista original:",Lista,"\n")


#Comando para añadir elemento a la lista:
Lista.append(pi)


print("2. Añadindo pi a la lista: ",Lista,"\n")
print("3. Suma de los elementos de la lista: ",sum(Lista),"\n")


#Segundo elemento de la lista:
elt=Lista[1]
Lista.pop(1)
print("4. Segundo elemento (",elt,") removido de la lista: ",Lista,"\n")


#Añadindo un nuevo elemento a la lista (digamos, elt):
Lista.insert(2,elt)
print("5: Elemento",elt,"añadido a la lista en la posicion 3:",Lista,"\n")


#Nova lista:
Lista2=[10,pi**4,sqrt(pi),log(0.3)]


#Concatenando listas:
Lista3=Lista+Lista2
print("6. Nueva lista creada:\n",Lista2,"\n")
print("Añadiendo a la lista antigua:\n",Lista3,"\n")


#creando lista ordenada
Lista4=sorted(Lista3)

print("7. Lista ordenada de menor a mayor:\n",Lista4,"\n")
#Lista ordenada de mayor a menor:
Lista5=sorted(Lista3,reverse=True)
print("8: Lista ordenada de mayor a menor:\n",Lista5)
```
