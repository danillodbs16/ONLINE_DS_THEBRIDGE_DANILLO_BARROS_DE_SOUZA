`````python
def encuentra_duplicados(lista):
    """Encuntra los elementos duplicados de una lista
    Entrada: lista de strings
    Salida: Índices de los strings duplicados"""
    #Verificar si todos los elementos en la lista son strings:
    #assert sum([isinstance(i,str) for i in lista])==len(lista), "Error! Ni todos los elementos de la lista son del tipo string."
    
    #Definir lista con strings en minusculas:
    lista_aux=[]
    for i in lista:
        try:
            lista_aux.append(i.lower())
        except:
            lista_aux.append(i)
    #lista_aux=[i.lower() for i in lista]
    elementos=set(lista_aux)

    #Crear dicionário con elementos duplicados (keys) y su lista de índices (values):
    Duplicados={}
    for e in elementos:
        
        contador=lista_aux.count(e)
        if contador>1:
            #Si el elemento se cuenta más de una vez:
            Duplicados[e]=[]
            indice=-1
            for elt in lista_aux:
                indice+=1
                if e==elt:
                    #Agregar indice a la lista de duplicados
                    Duplicados[e].append(indice)
    return Duplicados
      
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
def es_DNI(valor):
    letra=valor[-1]
    numeros=int(valor[:-1])
    palabra='TRWAGMYFPDXBNJZSQVHLCKE'
    letra_DNI=palabra[numeros%23]
    if 7<=len(valor)<=8:
        
        if letra_DNI==letra:
            
            result=True
        else:
            print("La letra del input no coincide con la letra de un DNI")
            result=False
    else:
        print("El input tiene menos de 7 dígitos o mas de 8 dígitos")
        result=False
    return result
`````
